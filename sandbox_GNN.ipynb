```python
import torch
from torch_geometric.datasets import ZINC
import networkx as nx
import matplotlib.pyplot as plt
```

```python
dataset = ZINC('/content/Graph-Neural-Networks/data')
```

```python
ex = dataset[0]
ex, ex['x']
```

```python
G = nx.Graph()

# Adding nodes to the NetworkX graph
for i in range(ex.num_nodes):
    G.add_node(i)

# Adding edges to the NetworkX graph
edge_index = ex.edge_index.t().numpy()
edges = [(edge[0], edge[1]) for edge in edge_index]
G.add_edges_from(edges)

# Visualize the graph
plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True, node_color='lightblue', font_weight='bold')
plt.title('ZINC Dataset: Example Graph Visualization')
plt.show()
```
